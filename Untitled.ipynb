```python
#Open the file and grab the histograms
fn = '/eos/user/c/cfreer/Moriond_2020/CMSSW_10_2_13/src/MonoZFinalFit/cards-ADD_MD_1_d_2/shapes-chBSM2018.root'
f = uproot.open(fn)
histograms = f.allitems( filterclass=lambda cls: issubclass(cls, uproot_methods.classes.TH1.Methods))
```

```python
#grab the histograms
for name, roothist in histograms:
    name = name.decode("utf-8")
    name = name.replace(";1", "")
    era = "2018"
    if  "ZZ" in name:
            ZZ_freq = np.abs(roothist.numpy())[0] 
            ZZ_0Jet = ZZ_freq[:len(ZZ_freq)//2]
            ZZ_1Jet = ZZ_freq[len(ZZ_freq)//2+1:]
    if  "WZ" in name:
            WZ_freq = np.abs(roothist.numpy())[0]
            WZ_0Jet = WZ_freq[:len(ZZ_freq)//2]
            WZ_1Jet = WZ_freq[len(ZZ_freq)//2+1:]
    if  "WW" in name:
            WW_freq = np.abs(roothist.numpy())[0]
            WW_0Jet = WW_freq[:len(ZZ_freq)//2]
            WW_1Jet = WW_freq[len(ZZ_freq)//2+1:]
    if  "TOP" in name:
            TOP_freq = np.abs(roothist.numpy())[0]
            TOP_0Jet = TOP_freq[:len(ZZ_freq)//2]
            TOP_1Jet = TOP_freq[len(ZZ_freq)//2+1:]
    if  "DY" in name:
            DY_freq = np.abs(roothist.numpy())[0]
            DY_0Jet = DY_freq[:len(ZZ_freq)//2]
            DY_1Jet = DY_freq[len(ZZ_freq)//2+1:]
    if  "VVV" in name:
            VVV_freq = np.abs(roothist.numpy())[0]
            VVV_0Jet = VVV_freq[:len(ZZ_freq)//2]
            VVV_1Jet = VVV_freq[len(ZZ_freq)//2+1:]
    if  "total_background" in name:
            bg_freq = np.abs(roothist.numpy())[0]
            bg_0Jet = bg_freq[:len(ZZ_freq)//2]
            bg_1Jet = bg_freq[len(ZZ_freq)//2+1:]
    if  "data_obs" in name:
            data_freq = np.abs(roothist.numpy())[0]
            data_0Jet = data_freq[:len(ZZ_freq)//2]
            data_1Jet = data_freq[len(ZZ_freq)//2+1:]
```
