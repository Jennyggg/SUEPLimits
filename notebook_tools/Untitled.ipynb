```python
import os
import json
```

```python
with open("../config/xsections_SUEP.json", "r") as f:
    infile = json.load(f)
```

```python
effs = []
mss = []

for sample, sampledata in infile.items():
    ms = sample.split("mS")[1].split('.')[0]
    if float(ms) not in mss:
        mss.append(int(ms))
        effs.append(sampledata['kr'])
```

```python
print(mss)
print(effs)
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
sorted_args = np.argsort(mss)
effs = np.array(effs)[sorted_args]
mss = np.array(mss)[sorted_args]
```

```python
plt.plot(mss, effs)
plt.ylabel("gen efficiency")
plt.xlabel("mS")
```
